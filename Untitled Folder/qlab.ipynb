```python
import matplotlib.pyplot as plt
import math
```

```python
def qexp(x, q=0.):
    if q==0:
        return math.exp(x)
    return (1.+x*q)**(1./q)

def qlog(x, q=0.):
    if q==0:
        return math.log(x)
    return (x**q - 1.)/q

def qsigmoid(x, q=0):
    if q==0:
        return math.exp(x)/(math.exp(x) + math.exp(-x))
    if x <= -abs(1./q):
        return 0.
    if x >= abs(1./q):
        return 1.
    return qexp(x,q)/(qexp(x,q) + 1./qexp(x,q))

def fat_f(x, p=2):
    if x == 0:
        return 1.
    return (abs(x)**p+1)**(1./p) + x

def fat_sigmoid(x, p):
    v1 = fat_f(x, p)
    v2 = fat_f(-x, p)
    return v1/(v1+v2)
```

```python
pqs = [0.1, 0.5, 1, 2, 10]
nqs = [-10, -2, -1, -.5, -.1]
qs = pqs + nqs + [0.]

min_x = -10
max_x = 10
X = [min_x + (max_x-min_x)*float(i)/1000. for i in range(1000+1)]

Y_logistic = [qsigmoid(x,0) for x in X]
Ys = [[qsigmoid(x,q) for x in X] for q in [10, 1, 2, .5, .1, -2]]

#for Y in Ys:
#    plt.plot(X, Y)
plt.plot(X, Y_logistic)
plt.plot(X, [fat_sigmoid(x, 3) for x in X], color='black')


plt.show()

plt.plot(X, [Y_logistic[i]-fat_sigmoid(X[i], 3) for i in range(len(X))])
plt.show()
```

```python
def y_opt(y, q=0):
    if q==1.:
        return .5
    if y >= 1.:
        return 1.
    if y <= 0.:
        return 0.
    
    r = (1.-y)/y
    p = 1./(q - 1.)
    return r**p / (1. + r**p)
```

```python
X = [float(i)/1000. for i in range(1, 1000)]
Ys = [[y_opt(x, q) for x in X] for q in [0, -1, -2, -.5]]
for Y in Ys:
    plt.plot(X, Y)
plt.show()
```

```python
def BqCE(Y, y_pred, q=0., lam=1.):
    sum_err = 0.
    for y in Y:
        sum_err -= y*qlog(y_pred, q)
        sum_err -= (1.-y)*qlog(1.-y_pred, q)
        if q != -1:
            sum_err += lam*(y_pred**(q+1.) + (1.-y_pred)**(q+1.))/(q+1.)
        else:
            sum_err += lam*(math.log(y_pred) + math.log(1.-y_pred))
    return sum_err

def SquareError(Y, y_pred, q, lam):
    sum_err = 0.
    for y in Y:
        sum_err += (y-y_pred)**2
    return sum_err

def predict_y(Y, q, lam, success_scale=1.1, fail_scale=-.8, loss=BqCE):
    y_pred = 0.5
    val = loss(Y, y_pred, q, lam)
    step = 0.02
    
    while abs(step) > 1e-12:
        t_pred = y_pred + step
        if t_pred <= 0. or t_pred >= 1.:
            step *= fail_scale
            continue
            
        tval = loss(Y, t_pred, q, lam)
        if tval < val:
            y_pred = t_pred
            val = tval
            #print y_pred, tval
            step *= success_scale
        else:
            step *= fail_scale
            
    return y_pred
```

```python
def reverse_iter(a):
    i = len(a)-1
    while i >= 0:
        yield a[i]
        i -= 1
```
